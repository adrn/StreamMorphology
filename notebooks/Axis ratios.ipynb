```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def p_ab_ac(ab, ac):
    ab = np.atleast_1d(ab)
    
    if ac >= 0.5:
        rmin = ac
    else:
        rmin = 0.5
    
    ret = np.zeros_like(ab)
    ret[ab>=rmin] = 3./(2*(1-rmin)) * (1 - ((2*ab[ab>=rmin] - 1 - rmin) / (1 - rmin))**2)
    
    return ret
```

```python
p_ab = p_ab_ac(np.linspace(0,1.,100), 0.55)
plt.plot(np.linspace(0,1.,100), p_ab)
plt.xlim(0.5,1.)
plt.axvline(0.77)
```
