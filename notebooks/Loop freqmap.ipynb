```python
from __future__ import division, print_function

# Std lib
import glob

# Third-party
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# Custom
import gary.dynamics as gd
import gary.integrate as gi
import gary.io as io
import gary.potential as gp
from gary.units import galactic
```

```python
w0 = np.load("/Users/adrian/projects/morphology/output/loop_E-0.1/w0.npy")
len(w0)
```

```python
potential = gp.LeeSutoTriaxialNFWPotential(v_c=0.22, r_s=30.,
                                           a=1., b=0.9, c=0.7, units=galactic)
```

```python
w0_subsample = w0[np.random.randint(len(w0),size=10)]
```

```python
t,w = potential.integrate_orbit(w0_subsample, dt=2., nsteps=32000, 
                                Integrator=gi.DOPRI853Integrator)
```

```python
for i in range(w.shape[1]):
    E = potential.total_energy(w[:,i,:3],w[:,i,3:])
    dE = np.abs(E[1:] - E[0])
    print(dE.max())
```

```python
nfiles = len(glob.glob("/Users/adrian/projects/morphology/output/loop_E-0.1/all/*.npy"))
nfiles
```

```python
all_freqs = []
for i in range(nfiles):
    filename = "/Users/adrian/projects/morphology/output/loop_E/all/{}.npy".format(i)
    try:
        f = np.load(filename)
    except:
        print("{} doesn't exist.".format(filename))
        all_freqs.append((f*np.nan).tolist())
    all_freqs.append(f.tolist())
all_freqs = np.array(all_freqs)
```

```python
all_freqs.shape
```

### Energy conservation

```python
dE = all_freqs[:,6,0]
plt.hist(np.log10(dE[np.isfinite(dE)]), log=True);
```

```python
freq_diff = (all_freqs[:,:6,1] - all_freqs[:,:6,0]) / all_freqs[:,:6,0]
freq_diff = np.abs(freq_diff).max(axis=1)
plt.hist(np.log10(freq_diff[np.isfinite(freq_diff)]));
```

```python
good_ix = np.isfinite(freq_diff) & (all_freqs[:,6,0] < 1E-10) & (all_freqs[:,6,1] < 1E-10)
good_freqs = all_freqs[good_ix,:6,0]
good_freqs.shape
```

```python
fig,axes = plt.subplots(1,3,figsize=(16,5))

# initial condition / diffusion rate
pt = axes[0].scatter(w0[good_ix,0], w0[good_ix,2],
                     c=np.log10(freq_diff[good_ix]), 
                     s=(np.log10(freq_diff[good_ix])+8)*10, 
                     marker='.', cmap='Greys', vmin=-6, vmax=-3)
# fig.colorbar(pt)
axes[0].set_xlim(-1.,75)
axes[0].set_ylim(-1.,75)
axes[0].set_xlabel(r'$x_0$ $[{\rm kpc}]$')
axes[0].set_ylabel(r'$z_0$ $[{\rm kpc}]$')

# loopy frequencies
ix = 1
axes[ix].plot(good_freqs[:,4]/good_freqs[:,3], good_freqs[:,2]/good_freqs[:,3], 
             linestyle='none', marker='.', alpha=0.4)
axes[ix].set_xlim(0.45,0.77)
axes[ix].set_ylim(0.45,0.77)
# axes[ix].set_xlim(0.45,0.95)
# axes[ix].set_ylim(0.45,0.95)
axes[ix].set_xlabel(r'$\Omega_\phi/\Omega_R$')
axes[ix].set_ylabel(r'$\Omega_z/\Omega_R$')

x = np.linspace(0,1,1000)
for n,m,l in [(4,-2,-1),(1,-1,0),(0,-2,1)]:
    line, = axes[ix].plot(x, -(n/m*x + l/m), marker=None, 
                          linestyle='--', color='#2166AC')
# axes[ix].plot(x, (-m*x-n)/l, marker=None, linestyle='--')
# axes[ix].plot(x, 2*x-0.5, marker=None, linestyle='--')

# cartesian frequencies
ix = 2
axes[ix].plot(good_freqs[:,0]/good_freqs[:,2], good_freqs[:,1]/good_freqs[:,2], 
             linestyle='none', marker='.', alpha=0.4)
axes[ix].set_xlim(0.3,1.2)
axes[ix].set_ylim(0.3,1.2)
axes[ix].set_xlabel(r'$\Omega_x/\Omega_z$')
axes[ix].set_ylabel(r'$\Omega_y/\Omega_z$')

# x = np.linspace(0,1,1000)
# n,m,l = (0,1,-1)
# axes[ix].plot(x, (-l*x-n)/m, marker=None, linestyle='--')

fig.tight_layout()
```
