```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import argrelmin, argrelmax
%matplotlib inline
```

```python
t = np.linspace(-10,-1,100)
w1 = 0.01*(t+5.)**2 + 1.
w2 = -0.01*(t+5.)**2 + 1.
w3 = -(0.02*(t+5)**3 + 0.09*(t+5.)**2 + 5.)
```

```python
def tuneNonPrime(n):
    # utility function to ensure that n does not have large prime divisors.
    #    if needed, decrease it (in the FFT primes are great evil).

    MAX_FFT_PRIME = 41

    m = n
    f = 2
    while (f**2 <= m):
        if m % f == 0:
            m /= f
        else:
            f += 1

    if m >= MAX_FFT_PRIME and n >= MAX_FFT_PRIME:
        n -= 1
        n = tuneNonPrime(n)

    return n
```

```python
def curve_up(w):
    if (w[0] > w.min()) and (w[-1] > w.min()):
        pass
    
    elif (w[0] < w.max()) and (w[-1] < w.max()):
        return -np.abs(w)
    
    elif ((w.argmax() == 0) or (w.argmax() == (len(w)-1)) or
          (w.armin() == 0) or (w.argmin() == (len(w)-1))):
        return np.abs(w)
        
    return w
```

```python
plt.figure()
plt.plot(t,w1)
plt.plot(t,curve_up(w1))

plt.figure()
plt.plot(t,w2)
plt.plot(t,curve_up(w2))

plt.figure()
plt.plot(t,w3)
plt.plot(t,curve_up(w3))
```
